The dataset file can be downloaded from huggingface: https://huggingface.co/lingxusb/TXpredict/resolve/main/TXpredictDB_v1.csv?download=true

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# ------------------------------------------------
# 1. Read the CSV containing Folder, Filename, Length
# ------------------------------------------------
csv_path = 'C:/Work/CellState/20250121_TXpredictDB/TXpredictDB_v1.csv'  # Change to your CSV file
df = pd.read_csv(csv_path)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns_palette = sns.color_palette("tab10")

map_taxa_to_color = {
    'bacteria': sns_palette[0],
    'archaea':  sns_palette[1],
    'fungi':    sns_palette[2]
}

taxa_colors = df['taxa'].map(map_taxa_to_color)

fig, ax = plt.subplots(figsize=(5.5, 5.5), dpi=600, layout='constrained')
scatter = ax.scatter(
    df['x'].values,
    df['y'].values,
    s=0.2,
    c=taxa_colors,
    marker='.',
    linewidths=None,
    ec='None',
    alpha=0.8,
    rasterized=True
)

ax.axis('equal')
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')

# ------------------------------------------------
# 7. Save and show
# ------------------------------------------------
plt.show()
```
